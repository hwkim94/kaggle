## Video
1. video(mp4) -> frame(ppm)
    - cd ./style_transfer_dir
    - mkdir ./input_video
    - mv ./{mp4 file} ./input_video
    - mkdir ./video_helper/City_of_Stars
    - ffmpeg -v quiet -i ./input_video/{mp4 file} ./video_helper/City_of_Stars/frame_%04d.ppm
2. frame(ppm)으로 부터 optical flow(flo) 계산
    - cd  ./video_helper
    - bash make-opt-flow.sh ./City_of_Stars/frame_%04d.ppm  ./City_of_Stars
3. ppm과 flo를 모두 google drive로 옮김
4. ppm과 flo를 이용하여 GPU를 통해 style transfer
